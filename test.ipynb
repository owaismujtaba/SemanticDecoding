```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QPushButton, QLabel, QVBoxLayout, QWidget
from PyQt5.QtMultimedia import QMediaPlayer, QMediaContent
from PyQt5.QtCore import QUrl

class AudioPlayerApp(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()
        self.mediaPlayer = QMediaPlayer()

    def initUI(self):
        # Main widget
        self.main_widget = QWidget()
        self.setCentralWidget(self.main_widget)
        
        # Layout
        layout = QVBoxLayout()
        self.main_widget.setLayout(layout)

        # Buttons
        self.play_button = QPushButton("Play Audio")
        self.play_button.clicked.connect(self.play_audio)
        layout.addWidget(self.play_button)

        # Status label
        self.status_label = QLabel("")
        layout.addWidget(self.status_label)

        # Window settings
        self.setWindowTitle('Audio Player')
        self.setGeometry(100, 100, 300, 100)
        self.show()

    def play_audio(self):
        media_content = QMediaContent(QUrl.fromLocalFile("/home/owais/GitHub/21-05-2024/EEGAnotator/audio_file.wav"))
        self.mediaPlayer.setMedia(media_content)
        self.mediaPlayer.play()

        # Check for errors
        self.mediaPlayer.errorOccurred.connect(self.handle_error)

    def handle_error(self, error):
        if error == QMediaPlayer.NoError:
            self.status_label.setText("Audio playback successful")
        else:
            self.status_label.setText("Error: {}".format(self.mediaPlayer.errorString()))

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = AudioPlayerApp()
    sys.exit(app.exec_())
```

```python
def find_closet_starting_point_in_eeg(eeg_events, timestamp):
    eeg_timestamps = [item[1] for item in eeg_events]
    diff = abs(eeg_timestamps-timestamp)
    close_index = np.argmin(diff) 
    return close_index
```

```python
def map_eeg_actions_marker_words(start_timestamp_eeg, eeg_events, markers_words_timestamps):
    word_index = 0
    result = []
    for event in eeg_events[start_timestamp_eeg:]:
        action , start_time, end_time, start_index, end_index ,duration = event

        for index in range(word_index, len(markers_words_timestamps)):
            marker, word, time = markers_words_timestamps[index]
            

            if action == marker:
                result.append([marker, word, start_time, end_time,
                               start_index, end_index, duration, time])
                word_index = index
                break

    return result
```
